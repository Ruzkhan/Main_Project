# 1)Outliers Handling

```python
import numpy as np
from scipy.stats import zscore
```

```python
features = data[['Fresh','Milk','Grocery','Frozen','Detergents_Paper','Delicassen']]
```

```python
# Using zscore to remove outliers
z=np.abs(zscore(features))
z
```

```python
# Creating new dataframe
new_df = data[(z<3).all(axis=1)] 
new_df
```

```python
# Shape of original dataset
data.shape
```

```python
#After Removing Outliers
new_df.shape
```

```python
# Checking the the data loss
data_loss = (440-414)/440*100
data_loss
```

Here we loss 2.170 % Data

# ii) IQR (Inter Quantile Range) method

```python
# 1st quantile
Q1=features.quantile(0.25)

# 3rd quantile
Q3=features.quantile(0.75)

# IQR
IQR=Q3 - Q1

df1=data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
#After removing outlier
df1.shape
```

```python
#dataset
X =data[['Channel', 'Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper',
       'Delicassen']]
X.head()
```

```python
#Elbow method
from sklearn.cluster import KMeans

wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
data.shape
```

```python
data.head()
```

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 4, random_state = 42)
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)
```

# Model Evaluation

```python
# Just like in supervised learning we have a metric in unsupervised learning as well
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X,y_kmeans)
```

```python
import matplotlib.pyplot as plt

# Visualising the clusters
plt.figure(figsize=(8, 5))

# Assuming X is a DataFrame with columns 'Channel', 'Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen'
# y_kmeans is a variable containing cluster assignments for each data point

# Scatter plot for Cluster 1 (assuming y_kmeans == 0 for Cluster 1)
plt.scatter(X[y_kmeans == 0]['Channel'], X[y_kmeans == 0]['Fresh'], s=100, c='red', label='Cluster 1')

# Scatter plot for Cluster 2 (assuming y_kmeans == 1 for Cluster 2)
plt.scatter(X[y_kmeans == 1]['Channel'], X[y_kmeans == 1]['Fresh'], s=100, c='blue', label='Cluster 2')

# Scatter plot for Cluster 3 (assuming y_kmeans == 2 for Cluster 3)
plt.scatter(X[y_kmeans == 2]['Channel'], X[y_kmeans == 2]['Fresh'], s=100, c='green', label='Cluster 3')

# Scatter plot for Cluster 4 (assuming y_kmeans == 3 for Cluster 4)
plt.scatter(X[y_kmeans == 3]['Channel'], X[y_kmeans == 3]['Fresh'], s=100, c='cyan', label='Cluster 4')



# Plotting cluster centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='black', label='Centroids')

plt.title('Clusters of customers')
plt.xlabel('Channel')
plt.ylabel('Fresh')
plt.legend()
plt.show()
```
